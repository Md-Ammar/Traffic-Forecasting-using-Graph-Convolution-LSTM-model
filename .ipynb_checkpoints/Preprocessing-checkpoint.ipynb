```python
data = pd.read_csv("distance(from-to-cost).csv")
```

```python
len(data)
```

```python
arr = np.array([[0.0 for i in range(170)] for j in range(170)])
```

```python
for i in range(len(data)):
    if arr[data['from'][i]][data['to'][i]] == 1:
        print(f"duplicate found {i}")
    arr[data['from'][i]][data['to'][i]] = arr[data['cost'][i]]/1000
    
    

arr
```

```python
dt = pd.DataFrame(arr)
```
